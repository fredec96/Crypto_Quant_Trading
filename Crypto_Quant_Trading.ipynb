````python
# Create BTC 10-day trading average and 20-day trading average dataframe
# Create BTC DataFrame
BTC_path = Path("../Crypto_Quant_Trading/Raw_CSV/BTC.csv")
BTC_df = pd.read_csv(BTC_path,
                    parse_dates=['date'],
                    index_col="date"
)

#Print BTC_df
BTC_df

````

````python
# Prepare for the analysis by converting the df for BTC prices to daily returns
daily_return = BTC_df['close'].pct_change()
movin_ave_df = BTC_df[['close']].copy()
movin_ave_df.insert(1, 'daily return', daily_return)
movin_ave_df = movin_ave_df.dropna()

# Insert '10day rolling average' and '20day rolling average'into movin_ave_df
# Print movin_ave_df
rolling_10_day = movin_ave_df['daily return'].rolling(window=10).mean()
rolling_20_day = movin_ave_df['daily return'].rolling(window=20).mean()
movin_ave_df.insert(2, '10day rolling average', rolling_10_day)
movin_ave_df.insert(3, '20day rolling average', rolling_20_day)


movin_ave_df
````

````python
# Filter and select the dates when 10day rolling average > 20day rolling average
# print first 5 rows of new dataframe
rolling10_greater_rolling20 = movin_ave_df[movin_ave_df["10day rolling average"] > 
                                           movin_ave_df["20day rolling average"]]


# Calculate cumulative returns for each window
cumulative_returns_10d = (1 + rolling10_greater_rolling20['10day rolling average']).cumprod() - 1
cumulative_returns_20d = (1 + rolling10_greater_rolling20['20day rolling average']).cumprod() - 1
rolling10_greater_rolling20.insert(3, '10day cumulative returns', cumulative_returns_10d)
rolling10_greater_rolling20.insert(5, '20day cumulative returns', cumulative_returns_20d)

rolling10_greater_rolling20
````
